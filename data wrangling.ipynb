```python
# Concatenate dataset_3 with df_combined
df_final = pd.concat([df_combined, df3], ignore_index=True)
```

```python
# Check missing values after merging
print("Missing Values After Merging:")
print(df_final.isnull().sum())
```

```python
# Ensure numeric columns before filling missing values
numeric_cols = df_final.select_dtypes(include=[np.number]).columns
df_final[numeric_cols] = df_final[numeric_cols].fillna(df_final[numeric_cols].mean())
```

```python
# Outlier detection using IQR
Q1 = df_final[numeric_cols].quantile(0.25)
Q3 = df_final[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Remove outliers
df_final = df_final[~((df_final[numeric_cols] < (Q1 - 1.5 * IQR)) | (df_final[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Save cleaned dataset
df_final.to_csv("cleaned_data.csv", index=False)
```
